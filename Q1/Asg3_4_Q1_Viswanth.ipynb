```python
import requests

url = "https://soccer-data.p.rapidapi.com/team/squad"

querystring = {"teamId":"803"}

headers = {
	"X-RapidAPI-Key": "cc404ccc41msh47d386923b9bc75p10733fjsn630bb24c9993",
	"X-RapidAPI-Host": "soccer-data.p.rapidapi.com"
}

response = requests.get(url, headers=headers, params=querystring)
api_data = response.json()
print(response.json())
```

```python
import pandas as pd


player_data = []
for entry in api_data:
    player = entry["player"]
    tournaments = entry["tournaments"]
    tournament_info = tournaments[0]["tournament"]
    season_info = tournaments[0]["season"]
    player_data.append(
        {
            "Name": player["knownName"],
            "Position": player["position"]["name"],
            "Country": player["country"]["name"],
            "Age": player["age"],
            "Tournament": tournament_info["name"],
            "Season": season_info["name"],
        }
    )

# Create a DataFrame
df = pd.DataFrame(player_data)

# Displaying the DataFrame
df.head()
```
